```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

import joblib
```

```python
# Data Historis
# [Suhu_Mesin, Getaran_Mesin]

X_train = np.array([
    [60,2],
    [70,4],
    [80,6],
    [90,8],
    [100,10]
])

# Target
y_train = np.array([
    10,
    25,
    45,
    70,
    95
])

# Standardisasi

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(
    X_train
)

# Training

model = LinearRegression()

model.fit(
    X_train_scaled,
    y_train
)

print("Model dan Scaler berhasil dilatih pada data historis.")
```

```python
joblib.dump(
    model,
    "model_risiko_v1.joblib"
)

joblib.dump(
    scaler,
    "scaler_risiko_v1.joblib"
)

print("Model berhasil disimpan.")
print("Scaler berhasil disimpan.")
```

```python
loaded_model = joblib.load(
    "model_risiko_v1.joblib"
)

loaded_scaler = joblib.load(
    "scaler_risiko_v1_.joblib"
)

data_baru = np.array([
    [85,7]
])

data_baru_scaled = loaded_scaler.transform(
    data_baru
)

hasil = loaded_model.predict(
    data_baru_scaled
)

print(
    "Prediksi Risiko:",
    hasil[0]
)
```

```python
def check_data_drift(
    new_data,
    train_mean,
    threshold=2.0
):

    drift = np.abs(
        np.mean(new_data) -
        np.mean(train_mean)
    )

    if drift > threshold:

        return (
            f"WARNING: Drift {drift:.2f}"
        )

    return "Sistem Stabil"

train_mean_ref = np.mean(
    X_train
)

data_anomali = np.array([
    [200,50]
])

print(
    check_data_drift(
        data_anomali,
        train_mean_ref
    )
)
```

```python
def clean_sensitive_data(
    df_input
):

    cols_remove = [
        "Nama_Operator",
        "NIK_Petugas"
    ]

    return df_input.drop(
        columns=[
            c
            for c in cols_remove
            if c in df_input.columns
        ]
    )

raw_input = pd.DataFrame({

    "Nama_Operator":[
        "Budi"
    ],

    "Suhu":[
        85
    ],

    "Getaran":[
        7
    ]
})

final_input = clean_sensitive_data(
    raw_input
)

print(final_input)
```
